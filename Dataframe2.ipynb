```python
import pandas as pd
import numpy as np
```

```python
data={'ID':[101,102,101,104,105,106,107,108,109,101],'Name':['Alice','Ben','Alice','John','Anil',np.nan,np.nan,np.nan,'Ray','Alice'],'Age':[45,42,45,34,36,39,44,28,19,45],
'Salary':[45000,78000,45000,630000,np.nan,98000,13000,56000,63000,45000]}
df=pd.DataFrame(data)
df
```

```python
df['Name'].fillna('Unknown')
```

```python
df['Salary'] = df['Salary'].fillna(df['Salary'].mean())
df
```

```python
duplicates=df[df.duplicated()]
duplicates
duplicates.count()
```

```python
remove_duplicates=df.drop_duplicates()
remove_duplicates
```
